```python
# For Creation of DataFrames
import pandas as pd
```

```python
# Login Credentials
import json
with open('/Users/aveld/secret/yelp_api.json') as f:
    login = json.load(f)
```

```python
# Yelp API Instance
from yelpapi import YelpAPI

yelp_api = YelpAPI(login['api-key'])
```

```python
# First page results, r for results
r = yelp_api.search_query(location = 'Boston, MA',
                         term = 'food')
# Verify dictionary in results
r.keys()
```

```python
# Check total results
r['total']
```

```python
# Check how many results in first page
len(r['businesses'])
```

```python
# Create first DataFrame results
biz_1 = pd.DataFrame(r['businesses'])
biz_1.head()
```

```python
# Obtaining Second Page of data, r for results
r = yelp_api.search_query(location = 'Boston, MA',
                         term = 'food',
                         offset = 20)
```

```python
# Create Second DataFrame
biz_2 = pd.DataFrame(r['businesses'])
biz_2.head()
```

```python
# Concatenate the two pages of results into 1 dataframe
biz_df = pd.concat([biz_1, biz_2],
                  ignore_index=True)
```
