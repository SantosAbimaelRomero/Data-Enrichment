# Exploring Data

## df_powers

```python
df_powers.duplicated().sum()
```

```python
df_powers.head()
```

I need to ultimately one-hot encode the Powers column

```python
df_powers.loc[1,'Powers']
```

```python
df1 = df_powers.copy()
df1['Powers'] = df1['Powers'].str.split(',')
```

```python
# The column should contain lists now
df1.head()
```

```python
ex = df1.explode('Powers')
```

```python
# Each value in the list is separated
ex
```

```python
cols = ex['Powers'].dropna().unique()
cols
```

```python
for m in cols:
    df_powers[m] = df_powers['Powers'].str.contains(m)
```

```python
df_powers.drop(columns='Powers', inplace=True)
```

```python
df_powers.head()
```

```python
df_powers.info()
```

## df_info

```python
df_info.info()
```

```python
df_info['Hero|Publisher']
```

I will separate hero and publisher into their own columns

```python
df_info[['Hero', 'Publisher']] = df_info['Hero|Publisher'].str.split('|', expand=True)
```

```python
df_info.head()
```

```python
df_info.drop(columns='Hero|Publisher', inplace=True)
```

```python
df_info.head()
```

I will separate height and weight into their own columns

```python
coord = df_info.loc[0,"Measurements"]
print(type(coord))
coord
```

```python
import json
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')

df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()
```

```python
measurements = df_info['Measurements'].apply(pd.Series)
measurements
```

```python
df_info = pd.concat([df_info, measurements], axis=1)
df_info.head()
```

```python
df_info.drop(columns='Measurements', inplace=True)
```

```python
df_info.head()
```

```python
df_info['Height'] = df_info['Height'].str.replace(' cm', '')
df_info[['Height']]
```

```python
df_info['Weight'] = df_info['Weight'].str.replace(' kg', '')
df_info[['Weight']]
```

```python
df_info.rename(columns={'Height':'Height (cm)', 'Weight':'Weight (kg)'}, inplace=True)

df_info.head()
```

```python
df_info.info()
```

```python
df_info['Height (cm)'] = df_info['Height (cm)'].astype(float)
df_info['Weight (kg)'] = df_info['Weight (kg)'].astype(float)
df_info.info()
```

# Final DataFrame

```python
df = pd.concat([df_info, df_powers], axis=1)
df.head()
```

```python
df.info()
```
